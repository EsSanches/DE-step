```python
sum_res = df.groupby(df['Дата заявки'].dt.month).agg({"Номер заявки": "count", "Сумма кредита": "sum"}).reset_index()
```

```python
sum_res['Сумма кредита'] = sum_res['Сумма кредита'].astype(np.int64)
```

```python
sum_res = sum_res.rename(columns={'Дата заявки': 'Месяц', 'Номер заявки': 'Кол-во заявок', 'Сумма кредита': 'Сумма выданых кредитов'})
```

```python
sum_res.to_excel('кол-во и сумма кредитов.xlsx', index=False)
```
